STD DEV ANALYSIS:
AB, SO, RA, ER have very high.

Considering the quartile and the data in them, spread of values is very large for AB, BB, SO,


```python
from numpy import percentile

    # calculate interquartile range
for column in df:
    q25, q75 = percentile(df[column], 25), percentile(df[column], 75)
    iqr = q75 - q25
    print('\n\nPercentiles for column %s are: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (column, q25, q75, iqr))

    # calculate the outlier cutoff
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off

    # identify outliers
    outliers = [x for x in df[column] if x < lower or x > upper]
    print('Identified number of outliers: %d' % len(outliers))

    # remove outliers
    outliers_removed = [x for x in df[column] if x >= lower and x <= upper]
    print('Non-outlier observations: %d' % len(outliers_removed))

    # Outliers
    print('Outlier values are : %s' %outliers)
```

R, ERA and some other columns have only one outlier value, we can ignore that. No other outliers found.
