```python
from numpy import percentile

    # calculate interquartile range
for column in df:
    q25, q75 = percentile(df[column], 25), percentile(df[column], 75)
    iqr = q75 - q25
    print('\n\nPercentiles for column %s are: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (column, q25, q75, iqr))

    # calculate the outlier cutoff
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off

    # identify outliers
    outliers = [x for x in df[column] if x < lower or x > upper]
    print('Identified number of outliers: %d' % len(outliers))

    # remove outliers
    outliers_removed = [x for x in df[column] if x >= lower and x <= upper]
    print('Non-outlier observations: %d' % len(outliers_removed))

    # Outliers
    print('Outlier values are : %s' %outliers)
```

### First we are trying to predict Average Price using Regression

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score,accuracy_score
```

```python
x=df_.drop(["AveragePrice"],axis=1)
x.head()
```

```python
y=df_['AveragePrice']
y.head()
```

```python
df_.info()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,2)==round(r2_score(y_test,pred_test)*100,2):
        print("Oprtimal Random state is : ",i)
        print("Training Score is : ",r2_score(y_train,pred_train)*100)
        print("Testing Score is : ",r2_score(y_test,pred_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=51)
```

### Linear Regression

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import Lasso

parameters = {'alpha':[0.0001,0.001,0.01,0.1,1,10], 'random_state':list(range(0,10))}
ls=Lasso()
clf= GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
ls= Lasso(alpha=0.0001, random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls = ls.predict(x_test)

lss = r2_score(y_test,pred_ls)
for j in range(2,10):
    lsscore = cross_val_score(ls,x,y,cv=j)
    lsc = lsscore.mean()
    print(" At CV :- ", j)
    print(":corss validation score is :- ", lsc*100)
    print("R2_score is :- ", lss*100)
    print("\n")
```

```python
ls.score(x_test,y_test)
```

```python
r2_score(y_test,pred_ls)
```

### Decission Tree

```python
from sklearn.tree import DecisionTreeRegressor

parameters= {'criterion':['mse','friedman_mse','mae'],'splitter':['best','random']}
dt=DecisionTreeRegressor()
clf=GridSearchCV(dt,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
dt = DecisionTreeRegressor(criterion='mae', splitter='random')
dt.fit(x_train,y_train)
dt.score(x_train,y_train)
```

```python
pred_decission = dt.predict(x_test)

dts= r2_score(y_test,pred_decission)
print("r2 Score :", dts*100)

dtscore= cross_val_score(dt,x,y,cv=3)
dtc = dtscore.mean()
print('Cross Validation score',dtc*100)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

parameters = {'criterion':['friedman_mse','mae'],'n_estimators':[100,200,300]}
rf=RandomForestRegressor()
clf= GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
rf = RandomForestRegressor(criterion='friedman_mse', n_estimators=100)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_random = rf.predict(x_test)

rfs= r2_score(y_test,pred_random)
print("r2 Score :", rfs*100)

rfscore= cross_val_score(rf,x,y,cv=3)
rfc = rfscore.mean()
print('Cross Validation score',rfc*100)
```

__Best model is Random Forest for Regression Problem. Hence, we will use it to save and for predictions.__
