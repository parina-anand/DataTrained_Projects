```python
from numpy import percentile

    # calculate interquartile range
for column in df:
    q25, q75 = percentile(df[column], 25), percentile(df[column], 75)
    iqr = q75 - q25
    print('\n\nPercentiles for column %s are: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (column, q25, q75, iqr))

    # calculate the outlier cutoff
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off

    # identify outliers
    outliers = [x for x in df[column] if x < lower or x > upper]
    print('Identified number of outliers: %d' % len(outliers))

    # remove outliers
    outliers_removed = [x for x in df[column] if x >= lower and x <= upper]
    print('Non-outlier observations: %d' % len(outliers_removed))

    # Outliers
    print('Outlier values are : %s' %outliers)
```

```python
# Changing Yes and No values of target column "Attrition" into 1 and 0.


change = {'Yes':1, 'No':0}
df['Attrition'] = df["Attrition"].apply(lambda x: change[x])
df.head()
```

```python
# Dropping categorical columns which have been encoded.

df.drop(["BusinessTravel","Department","EducationField","Gender","JobRole","MaritalStatus","OverTime"],axis=1,inplace=True)
```

```python
# Now concatinating encoded coulmns to dataset.

df=pd.concat([df,BusinessTravel_n,Department_n,EducationField_n,Gender_n,JobRole_n,MaritalStatus_n,OverTime_n],axis=1)
```

```python
# Now Dataset is ready for use
# Splitting train and test data
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
x=df.drop('Attrition',axis=1)
```

```python
x.head()
```

```python
y=df['Attrition']
```

```python
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr=LogisticRegression()
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,4)==round(accuracy_score(y_test,pred_test)*100,4):
        print("Best suitable Random state is: ",i)
        print("training Accuracy is: ",accuracy_score(y_train,pred_train)*100)
        print("testing accuracy is: ",accuracy_score(y_test,pred_test)*100)
```

```python
# We are taking Random State as 81

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=81,test_size=.020)
```

### Logistic Regression

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

### Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
grid_param={'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter' : ['best', 'random']}
```

```python
grid_search=GridSearchCV(dt,grid_param,cv=5)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_param=grid_search.best_params_
print(best_param)
```

```python
dtn= DecisionTreeClassifier(criterion='gini',max_depth=3,min_samples_leaf=9,min_samples_split=5,splitter='random')
```

```python
dtn.fit(x_train,y_train)
dtn.score(x_test,y_test)
```
